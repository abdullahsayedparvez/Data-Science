```python
stock = pd.read_csv('google_pytrend_MovingAverage.csv')
```

```python
stock.drop(['30DayAvg_Combined' , 'Combined_Index' , 'Unnamed: 12' ] , axis = 1 , inplace = True)
stock.drop(['company'] , axis = 1 , inplace = True)
stock.drop(['company_sum'] , axis = 1 , inplace = True)
stock.rename(columns = {'Unnamed: 6' : 'recheck_avg'},inplace = True)
```

```python
stock = stock.loc[stock['symbol'] == 'ABBV']
```

```python
start = 0
end = 31
mean_row = 30
for i , row in stock.iterrows():
    thirty_rows = stock.iloc[start:end]
    mean_temp_stock = thirty_rows['symbol_sum'].iloc[:-1].mean()
    stock.loc[mean_row,'recheck_avg'] = mean_temp_stock
    mean_row = mean_row + 1
    start = start + 1
    end = end + 1
    
    if end == 41:
        break
#     for i in range(30,40):
#         stock.loc[i,'recheck_avg'] = mean_temp_stock
```
