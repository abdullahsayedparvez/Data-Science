# MATPLOTLIB

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3] , [2,5,7])
plt.show()
```

```python
plt.plot([1,2,3] , [2,5,7])
plt.title('Abdullah is DATA')
plt.ylabel('This is Y')
plt.xlabel('This is X')
plt.show()
```

```python
x = [1,5,4]
y = [23,45,89]
x2 = [3,7,5]
y2 = [5,8,9]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
#style
from matplotlib import style
style.use('ggplot')
x = [1,5,4]
y = [23,45,89]
x2 = [3,7,5]
y2 = [5,8,9]
plt.plot(x,y,label = 'First' , linewidth = 4)
plt.plot(x2,y2 , label = 'Second')
plt.legend()
plt.show()

```

```python
x = [1,5,4]
y = [23,45,89]
x2 = [3,7,5]
y2 = [5,8,9]
plt.scatter(x,y,label = 'First' , linewidth = 4)
plt.scatter(x2,y2 , label = 'Second')
plt.legend()
plt.show()
```
