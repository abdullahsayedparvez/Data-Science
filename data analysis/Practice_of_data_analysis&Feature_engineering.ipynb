```python
dataset = pd.read_csv('bike_data.csv')
```

# A new column for years

```python
dataset['Year'] = dataset['Name'].str.extract('(\d\d\d\d)',expand=True)
```

# Removing a Year from a Name column

```python
dataset['Name'] = dataset['Name'].str.replace('(\d\d\d\d)' , '')
```

# New column for Power Bike

```python
dataset['Power_bike'] = dataset['Name'].str.extract('(\d\d\d)')
```

# Removing the bike power from Name column

```python
dataset['Name'] = dataset['Name'].str.replace('(\d\d\d)' , '')
```

# New  Column for different bike column

```python
dataset['different_bike_power(cc)'] = dataset['Name'].str.extract("(cc){1}")
```

# Removing bike_power from Name columns

```python
dataset['Name'] = dataset['Name'].str.replace("[cc{1}]" , '')
```

# NAN Values

```python
feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes == 'O']
print(feature_nan)
def replace_cat_feature(dataset,feature_nan):
    data = dataset.copy()
    data[feature_nan] = data[feature_nan].fillna('missing')
    return data
dataset = replace_cat_feature(dataset,feature_nan)

dataset[feature_nan].isnull().sum()
```
